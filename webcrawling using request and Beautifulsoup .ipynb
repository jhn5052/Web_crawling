```python
import requests
from bs4 import BeautifulSoup
import json
import os

## python 파일 위치
BASE_DIR = os.path.dirname(os.path.realpath('__file__'))

## HTTP GET Request
req = requests.get('https://movie.naver.com/movie/sdb/rank/rmovie.nhn')
```

```python
## HTML 소스 가져오기
html = req.text
## beautifulsoup로 html소스를 python객체로 변환
## 첫 인자는 html소스코드, 두번째 인자는 어떤 parser을 이용할지 명시
## 여기에서는 Python 내장 html.parser 이용
soup = BeautifulSoup(html, 'html.parser')
movie_titles = soup.select(
    '#old_content > table > tbody > tr > td.title > div > a'
)

data = {}

for title in movie_titles:
    ## tag 안의 텍스트
    print(title.text)
    
    ## tag 속성 갖고오기
    print(title.get('href'))

for title in movie_titles:
    data[title.text] = title.get('href')

with open(os.path.join(BASE_DIR, 'result.json'), 'w+') as json_file:
    json.dump(data, json_file)
```
